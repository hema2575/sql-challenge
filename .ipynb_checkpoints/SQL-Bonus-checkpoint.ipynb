```python
employees_db = pd.read_sql('select * from employees', connection)
departments_db = pd.read_sql('select * from departments', connection)
dept_emp_db = pd.read_sql('select * from dept_emp', connection)
dept_manager_db = pd.read_sql('select * from dept_manager', connection)
salaries_db = pd.read_sql('select * from salaries', connection)
titles_db = pd.read_sql('select * from titles', connection)
```

```python
salemp_db = pd.merge(salaries_db, employees_db, how = 'inner', on='emp_no')
salemp_db.head()
```

```python
salemp_db.rename(columns = {'emp_title_id' : 'title_id'}, inplace = True)
salemp_db.head()
```

```python
salemptitle_db = pd.merge(titles_db, salemp_db, how = 'inner', on='title_id')
```

```python
salemptitle_db
```

Histogram of the most common salary ranges for employees

```python
%matplotlib inline 
salary_emp = salemptitle_db['salary']

# Characterize the data set using matplotlib and stats.normaltest
plt.figure(figsize=(10,5),facecolor='white',edgecolor='blue')
plt.hist(salary_emp)

plt.xlabel('Salary')
plt.ylabel('Counts')
plt.show()
print(sts.normaltest(salemptitle_db['salary'].sample(100000)))
print("Mean salary :", salemptitle_db['salary'].mean())
print("Median of the salaries :", salemptitle_db['salary'].median())
print("Mode of the salaries :", salemptitle_db['salary'].mode())
```

```python
'''
We find that most staff have salaries in the range of $40000 to $50000.  Fewest employees are paid in the range $100,000 to $120,000. 
Also from the p value resulting from the normality test, the salaries do not appear to be normally distributed

'''
```

```python
salemptitle = salemptitle_db.groupby(['title']).mean()['salary'].round(3)

salemptitle_df = pd.DataFrame(salemptitle)
salemptitle_r_df = salemptitle_df.reset_index()
salemptitle_r_df
```

*Barchart of average salary by Title*

```python
%matplotlib inline 

x_axis = np.arange(len(salemptitle_r_df))
tick_locations = [value for value in x_axis]

# Set x and y limits
#plt.xlim(-0.75, len(x_axis))
#plt.ylim(0, max(salemptitle_r_df["salary"])+20)

#plt.figure(figsize=(15,10))
#plt.bar(x_axis, salemptitle_r_df["salary"], color='r', alpha=0.5, align="center")
salemptitle_r_df.plot(kind='bar',figsize=(10,5))
plt.xticks(tick_locations, salemptitle_r_df["title"], rotation="vertical")


plt.title("Average salary by title")
plt.xlabel("Titles")
plt.ylabel("Average Salary")
```

```python
'''
The average salaries of employees appear to be in the just less than $60,000. We see high mean salaries for titles Senior Staff and Staff. 
The lowest mean salary is just less than $50,000 and is drawn by titles Engineer, Assistant Engineer, Senior Engineer, Technique Leader. Followed by
a slight increase in salary for Manager. 
This gives ample proof to suspect that the data provided is not indicative of real world scenario. 

'''
```
